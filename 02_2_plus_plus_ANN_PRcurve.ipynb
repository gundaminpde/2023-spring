# **0. 전처리**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

### from sklearn.neighbors import KNeighborsClassifier

Url_B_cancer="https://raw.githubusercontent.com/gundaminpde/2023-spring/main/Dset_Breast%20Cancer_Wisconsin.csv"


df = pd.read_csv(Url_B_cancer)

df.head()
```

Attribute Information:

* Diagnosis (M = malignant, B = benign)



```python
df.info()
```

```python
df=df.drop(['id',"Unnamed: 32"],axis=1)
```

```python
df.info()
```

```python
y_original = df['diagnosis']
```

```python
y_original.head()
```

```python
y_original_2=pd.get_dummies(y_original)

y_original_2.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:,1:], y_original_2['B'],stratify=y_original_2['B'], random_state=3)
```

```python
X_train.head()
```

```python
std = StandardScaler()
std.fit(X_train.iloc[:,0:])
X_train.iloc[:,0:]= std.transform(X_train.iloc[:,0:])
X_train.head()
```

```python
X_test.iloc[:,0:]= std.transform(X_test.iloc[:,0:])
```

# **1. 기본적인 모델**

```python
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense    


model = Sequential()                                                  # 딥러닝 모델의 구조를 결정합니다.
model.add(Dense(30, input_dim=30, activation='relu'))
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])  # 딥러닝 모델을 실행합니다.

history=model.fit(X_train, y_train, epochs=70, batch_size=16)

```

```python
y_loss = history.history['loss']
y_accu = history.history['binary_accuracy']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
plt.plot(x_len,y_accu, marker='.', c="red")

plt.show()
```

```python
results_pred = model.predict(X_test)

from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_test, results_pred)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):

    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16) # Not shown in the book
    plt.xlabel("Threshold", fontsize=16)        # Not shown
    plt.grid(True)                              # Not shown
    plt.axis([0,1, 0, 1])             # Not shown


plt.figure(figsize=(8, 4))                                                                  # Not shown
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)

plt.show()
```

```python
def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
#plt.plot([recall_90_precision, recall_90_precision], [0., 0.9], "r:")
#plt.plot([0.0, recall_90_precision], [0.9, 0.9], "r:")
#plt.plot([recall_90_precision], [0.9], "ro")
plt.show()
```

```python
close_zero = np.argmin(np.abs(thresholds))
plt.plot(precisions[close_zero], recalls[close_zero], 'o', markersize=10,
         label="임계값 0", fillstyle="none", c='k', mew=2)

plt.plot(precisions, recalls, label="P_R_curve")
plt.xlabel("P")
plt.ylabel("R")
plt.legend(loc="best")
plt.show() # 책에는 없음
```

```python
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense    


model = Sequential()                                                  # 딥러닝 모델의 구조를 결정합니다.
model.add(Dense(30, input_dim=30, activation='relu'))
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['binary_accuracy'])  # 딥러닝 모델을 실행합니다.

history=model.fit(X_train, y_train, epochs=4, batch_size=16)
```

```python
y_loss = history.history['loss']
y_accu = history.history['binary_accuracy']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
plt.plot(x_len,y_accu, marker='.', c="red")

plt.show()
```

```python
results_pred = model.predict(X_test)

from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_test, results_pred)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):

    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16) # Not shown in the book
    plt.xlabel("Threshold", fontsize=16)        # Not shown
    plt.grid(True)                              # Not shown
    plt.axis([0,1, 0, 1])             # Not shown


plt.figure(figsize=(8, 4))                                                                  # Not shown
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)

plt.show()
```

```python
def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
plt.show()
```

```python

from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, results_pred)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # 대각 점선
    plt.axis([0, 1, 0, 1])                                    # Not shown in the book
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) # Not shown
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    # Not shown
    plt.grid(True)                                            # Not shown

plt.figure(figsize=(8, 6))                                    # Not shown
plot_roc_curve(fpr, tpr)
plt.show()
```
